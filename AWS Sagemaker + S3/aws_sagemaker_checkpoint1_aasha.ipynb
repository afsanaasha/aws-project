```python
my_bucket = 'dsba-6190-aasha'
my_file = 'Magazine_Subscriptions.csv'
s3client = boto3.client('s3')
response = s3client.get_object(Bucket=my_bucket, Key=my_file)
body = response['Body']
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(body.read()))
df.head()
```

```python
df.count()
```

```python
magazine = df.drop_duplicates()
magazine.head()
```

```python
magazine["reviewText"].head(10)
```

```python
from sqlite3 import connect
conn = connect(':memory:')
magazine.to_sql('mag_subs', conn)
```

```python
pd.read_sql('SELECT overall, verified, reviewText FROM mag_subs', conn)
```

```python
pd.read_sql('SELECT count(*), overall FROM mag_subs GROUP BY overall ORDER BY overall DESC', conn)
```
